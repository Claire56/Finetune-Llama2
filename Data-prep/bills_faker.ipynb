# Generate Synthetic Data

```python
import json, pandas as pd
import random
import datetime
from faker import Faker

# Initialize Faker
fake = Faker()

# Define a list of possible services
services = [
    'In-patient stay',
    'Out-patient visit',
    'Surgery',
    'Physiotherapy',
    'Consultation',
    'Radiology',
    'Laboratory tests'
]

# Define a list of possible payment options
payment_options = [
    'Credit card',
    'Bank transfer',
    'Check',
    'Online payment'
]

# Define a function to generate a random bill statement
def generate_bill_statement():
    num_services = random.randint(1,  3)
    service_list = random.sample(services, num_services)
    total = sum(random.randint(100,  500) for _ in range(num_services))
    name = f" {fake.first_name()} {fake.last_name()}"
    instruction_type = random.choice(['bill', 'payment', 'services', 'date'])
    payment_options = random.choice(["using Credit Card or Debit card", "by Going to paymybillonlin.rev, or using credit card", " using Insurance , Cash or Check"])
    date = fake.date_between(start_date='-1y', end_date='today')
    return {
        'services': ', '.join(service_list),
        'name': f'{name}',
        'instruction_type': f'{instruction_type}',
        'payment_options': f'{payment_options}',
        'total': f'${total}',
        'date': f'{date}'
    }

# generate_bill_statement()

# Define a function to generate a random instruction
def generate_instruction(bill_statement):
    # instruction_type = random.choice(['bill', 'payment', 'services'])
    if bill_statement["instruction_type"] == 'bill':
        return f"What is the bill for {bill_statement['name']}"
    elif bill_statement["instruction_type"] == 'payment':
        return f"My name is {bill_statement['name']}. How can I pay my out standing bill"
    elif bill_statement["instruction_type"] == 'services':
        return f"What services did {bill_statement['name']} receive at the hospital"
    else:
        return f"What date did {bill_statement['name']} receive services at the hospital"

# Define a function to generate a random response
def generate_response(bill_statement):
    if bill_statement["instruction_type"] == 'bill':
        return f"The bill for {bill_statement['name']} is {bill_statement['total']}"
    elif bill_statement["instruction_type"] == 'payment':
        return f" You can pay your bill {bill_statement['payment_options']}. "
    elif bill_statement["instruction_type"] == 'services':
        return f"The services received by {bill_statement['name']} at the hospital were: {bill_statement['services']}"
    else:
        return f" {bill_statement['name']} visited the hospital on {bill_statement['date']}."

# Generate the dataset
dataset = []
for _ in range(5000):
    bill_statement = generate_bill_statement()
    instruction = generate_instruction(bill_statement)
    response = generate_response( bill_statement)
    context = f"{bill_statement['name']}, visited the hospital on {bill_statement['date']} and recieved the following Services: {bill_statement['services']} costing a Total of {bill_statement['total']}. The bill can be paid {bill_statement['payment_options']}"
    dataset.append({
        'instruction': instruction,
        'context': context,
        'response': response
    })

# Write the dataset to a JSONL file
with open('bills.jsonl', 'w') as f:
    for record in dataset:
        f.write(json.dumps(record) + '\n')
```
